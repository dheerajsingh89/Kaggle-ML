## initialising the convolutional neural network

```python
cnn=tf.keras.models.Sequential() #same as the initialisation of Artificial Neural Network here also we are initialising the convolutional neural network as a sequence model
```

## adding convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[80,80,3]))
```

## adding max pool

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
```

## adding another convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))#as in first convolutional layer we already entered the input_shape so in second layer we don't want to mention it again
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
```

## flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

## adding full connection

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

## output layer

```python
cnn.add(tf.keras.layers.Dense(units=43,activation='softmax'))
```

## compiling

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

## training set according to train set and validation set

```python
cnn.fit(x=training_set, validation_data=val_set, epochs=30)
```
